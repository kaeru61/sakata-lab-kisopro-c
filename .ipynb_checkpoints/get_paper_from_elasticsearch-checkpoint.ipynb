```python
import pandas as pd
from elasticsearch import helpers
from elasticsearch import Elasticsearch
es = Elasticsearch(
    "http://133.11.72.200:9200/",
    # ca_certs="./ti06.crt",
    timeout=500,
    basic_auth=('oa_reader', 'five@godtruth')
)

def build_es_query_titleabstract_or(keywords,gte=1800,lte=2030):
    """Build Elasticsearch query for a specific keyword."""

    list_keywords = []
    for key in keywords:
        list_keywords = list_keywords + [  {
                        "match_phrase": {
                          "abstract": key
                        }
                      },
                                      {
                        "match_phrase": {
                          "title": key
                        }
                      }]
    
    return {
          "query": {
            "bool": {
              "must": [
                {
                  "range": {
                    "publication_year": {
                      "gte": gte,   
                      "lte": lte   
                    }
                  }
                },
                { "range" : { "cited_by_count" :{
                        "gte" : 0,
                    } }},   
                {
                  "bool": {
                    "should": list_keywords
                }
                }
              ]
            }
          }
        }
    

def build_es_query_topic(x):
    """Build Elasticsearch query for a topic."""
    topics = pd.read_pickle('/disks/qnap3/shared/openalex-24/data/master/topics.pickle')
    if x in topics.index:
        keyword = topics.loc[x].display_name
    else:
        keyword = topics.loc["https://openalex.org/"+str(x)].display_name
        
    return {
        "query": {
            "bool": {
            "must" : [
                { "match" : { "language" : "en" }},
                { "term" : { "topics": {"value":keyword} }}, 
                { "range" : { "cited_by_count" :{"gte" : 0,} }},       
                { "range" : { "publication_year" :{"gte" : 0,"lte":2025} }}
            ]

            }
        }
    }
def build_es_query_by_ids(id_list):
    return {
        "query": {
            "bool": {
                "must": [
                    {
                        "terms": {
                            "id": id_list
                        }
                    }
                ]
            }
        }
    }




def get_result(es_client, es_index, es_query, scroll_time="2m", batch_size=1000):
    all_hits = []
    response = es_client.search(index=es_index, body=es_query, scroll=scroll_time, size=batch_size)
    scroll_id = response['_scroll_id']
    
    while response['hits']['hits']:
        all_hits.extend(hit['_source'] for hit in response['hits']['hits'])
        response = es_client.scroll(scroll_id=scroll_id, scroll=scroll_time)
    
    es_client.clear_scroll(scroll_id=scroll_id)
    result = pd.DataFrame(all_hits)
    
    if 'cited_by_count' in result.columns:
        result = result.sort_values('cited_by_count')
    if 'id' in result.columns:
        result.index = result['id']
    
    journal = pd.read_pickle('/disks/qnap3/shared/openalex-24/data/journal/journal.pickle')
    result['journal'] = result['primary_location'].map(journal.get('display_name', {}))
    result['journal_if'] = result['primary_location'].map(journal.get('2yr_mean_citedness', {}))
    
    print(len(result))
    return result
```

```python
def get_generation_metadata(start_ids, es, generation, visited_ids):
    """
    start_ids: list of paper ids to start from
    generation: int, which generation this corresponds to
    visited_ids: set of already seen paper ids to prevent cycles
    """
    # 1. Elasticsearchで引用メタデータ取得
    query = build_es_query_by_ids(start_ids)
    ref_metadata = get_result(es, "oa_papers", query).reset_index(drop=True)

    # 2. 新たな reference ペアを展開
    reference_pairs = ref_metadata.dropna(subset=["reference"]).explode("reference")[["id", "reference"]]
    reference_pairs.columns = ["citing_paper_id", "referenced_paper_id"]

    # 3. すでに見た論文は除外（ループ防止）
    reference_pairs = reference_pairs[~reference_pairs["referenced_paper_id"].isin(visited_ids)]

    # 4. 各被引用論文に対する引用元IDをカンマ結合
    reference_summary = (
        reference_pairs
        .groupby("referenced_paper_id")["citing_paper_id"]
        .apply(lambda ids: ",".join(map(str, sorted(ids.unique()))))
        .reset_index()
    )
    reference_summary.columns = ["referenced_paper_id", "in_degree_ids"]

    # 5. 被引用論文のメタデータ取得
    next_ids = reference_summary["referenced_paper_id"].tolist()
    next_metadata = get_result(es, "oa_papers", build_es_query_by_ids(next_ids)).reset_index(drop=True)

    # 6. メタ情報に in_degree_ids をマージ
    result = next_metadata.merge(
        reference_summary,
        left_on="id",
        right_on="referenced_paper_id",
        how="left"
    )

    result["generation"] = generation

    return result, set(next_ids)
```

```python
def get_generation_metadata(start_ids, es, generation, visited_ids):
    """
    start_ids: list of paper ids to start from
    generation: int, which generation this corresponds to
    visited_ids: set of already seen paper ids to prevent cycles
    """
    drop_cols = ["abstract", "concepts", "primary_location", "doi"]

    # 1. Elasticsearchで引用メタデータ取得
    query = build_es_query_by_ids(start_ids)
    ref_metadata = get_result(es, "oa_papers", query).drop(columns=drop_cols, errors="ignore").reset_index(drop=True)

    # 2. 新たな reference ペアを展開
    reference_pairs = ref_metadata.dropna(subset=["reference"]).explode("reference")[["id", "reference"]]
    reference_pairs.columns = ["citing_paper_id", "referenced_paper_id"]

    # 3. すでに見た論文は除外（ループ防止）
    reference_pairs = reference_pairs[~reference_pairs["referenced_paper_id"].isin(visited_ids)]

    # 4. 各被引用論文に対する引用元IDをカンマ結合
    reference_summary = (
        reference_pairs
        .groupby("referenced_paper_id")["citing_paper_id"]
        .apply(lambda ids: ",".join(map(str, sorted(ids.unique()))))
        .reset_index()
    )
    reference_summary.columns = ["referenced_paper_id", "in_degree_ids"]

    # 5. 被引用論文のメタデータ取得（次世代のノード）
    next_ids = reference_summary["referenced_paper_id"].tolist()
    next_metadata = get_result(es, "oa_papers", build_es_query_by_ids(next_ids)).drop(columns=drop_cols, errors="ignore").reset_index(drop=True)

    # 6. メタ情報に in_degree_ids をマージ
    result = next_metadata.merge(
        reference_summary,
        left_on="id",
        right_on="referenced_paper_id",
        how="left"
    )

    result["generation"] = generation

    return result, set(next_ids)
```

```python
from IPython.display import display

def get_generation_metadata(start_ids, es, generation, visited_ids):
    """
    start_ids: list of paper ids to start from
    generation: int, which generation this corresponds to
    visited_ids: set of already seen paper ids to prevent cycles
    """
    drop_cols = ["abstract", "concepts", "primary_location", "doi"]

    print(f"\n==== Generation {generation} ====")
    print(f"クエリ対象ID数: {len(start_ids)}")

    # 1. Elasticsearchで引用メタデータ取得
    query = build_es_query_by_ids(start_ids)
    ref_metadata = get_result(es, "oa_papers", query).drop(columns=drop_cols, errors="ignore").reset_index(drop=True)

    print(f"メタデータ取得: {len(ref_metadata)} 論文")
    display(ref_metadata[["id", "title"]].head())

    # 2. 新たな reference ペアを展開
    reference_pairs = ref_metadata.dropna(subset=["reference"]).explode("reference")[["id", "reference"]]
    reference_pairs.columns = ["citing_paper_id", "referenced_paper_id"]

    print(f"展開された引用ペア数: {len(reference_pairs)}")
    display(reference_pairs.head())

    # 3. すでに見た論文は除外（ループ防止）
    reference_pairs = reference_pairs[~reference_pairs["referenced_paper_id"].isin(visited_ids)]

    print(f" 未訪問の引用先: {len(reference_pairs)}")
    display(reference_pairs.head())

    # 4. 各被引用論文に対する引用元IDをカンマ結合
    reference_summary = (
        reference_pairs
        .groupby("referenced_paper_id")["citing_paper_id"]
        .apply(lambda ids: ",".join(map(str, sorted(ids.unique()))))
        .reset_index()
    )
    reference_summary.columns = ["referenced_paper_id", "in_degree_ids"]

    print(f"被引用論文数（次世代）: {len(reference_summary)}")
    display(reference_summary.head())

    # 5. 被引用論文のメタデータ取得（次世代のノード）
    next_ids = reference_summary["referenced_paper_id"].tolist()
    next_metadata = get_result(es, "oa_papers", build_es_query_by_ids(next_ids)).drop(columns=drop_cols, errors="ignore").reset_index(drop=True)

    print(f"次世代メタデータ取得: {len(next_metadata)}")
    display(next_metadata[["id", "title"]].head())

    # 6. メタ情報に in_degree_ids をマージ
    result = next_metadata.merge(
        reference_summary,
        left_on="id",
        right_on="referenced_paper_id",
        how="left"
    )

    result["generation"] = generation

    print(f" Generation {generation} 完了: {len(result)} 論文")
    display(result[["id", "title", "in_degree_ids"]].head())

    return result, set(next_ids)
```
